```python
import pandas as pd
import numpy as np
```

# Data Load

```python
data = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
df = pd.read_csv(data)
```

# Data Preprocessing

```python
x = df.iloc[:,0:13]
y = df.iloc[:, -1]
```

```python
x["intercept"]=1
x = x[x.columns[::-1]]
del x['chas']
```

```python
x
```

```python
xx = x.to_numpy()
```

# QR Decomposition

```python
Q = []
a = xx.copy()
for i in range(np.shape(a)[1]):
    s = a[:,i]
    for j in range(i):
        s -= (np.dot(a[:,i],a[:,j])/np.sum(a[:,j]*a[:,j]))*a[:,j]
    Q.append(s/np.sqrt(np.sum(s*s)))
Q = np.transpose(Q)
```

```python
R = np.zeros(shape=(13,13))
for i in range(np.shape(a)[1]):
    for j in range(i+1):
        R_temp = np.dot(Q[:,j],xx[:,i])
        R[j,i] = R_temp
```

```python
A = Q@R
```

```python
pd.DataFrame(A)
```

# Least Square Method

```python
Y = np.transpose(Q)@y
```

```python
beta = np.zeros(np.shape(Y))

for i in range(13-1, -1, -1):
    tmp = Y[i]
    for j in range(13-1, i, -1):
        tmp -= beta[j]*R[i,j]
        
    beta[i] = tmp/R[i,i]
```
