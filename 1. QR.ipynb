# QR Decomposition

```python
Q = []
a = xx.copy()
for i in range(np.shape(a)[1]):
    s = a[:,i]
    for j in range(i):
        s -= (np.dot(a[:,i],a[:,j])/np.sum(a[:,j]*a[:,j]))*a[:,j]
    Q.append(s/np.sqrt(np.sum(s*s)))
Q = np.transpose(Q)
```

# Least Square Method

```python
Y = np.transpose(Q)@y
```

```python
beta = np.zeros(np.shape(Y))

for i in range(13-1, -1, -1):
    tmp = Y[i]
    for j in range(13-1, i, -1):
        tmp -= beta[j]*R[i,j]
        
    beta[i] = tmp/R[i,i]
```

# Comparison

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression(fit_intercept=False)
mlr.fit(x, y) 
```

# Inverse of R

```python
iden = np.identity(n=13)
```

```python
inv_R=np.zeros(shape=(13,13))
```

```python
for i in range(12,-1,-1):
    for j in range(12,-1,-1):
        inv_R[j,i] = (iden[i,j]-np.sum(R[j]*inv_R[:,i]))/R[j,j]
```

```python
np.matrix.round(inv_R@R,1)
```

# Standard Error of $\hat{\beta}$ and p-value

```python
y_hat = xx@beta
mse = np.sum((y_hat-y)**2)/(506-13)
variance_of_beta_hat = (inv_R@np.transpose(inv_R))*mse
se = np.sqrt(np.diag(variance_of_beta_hat))
```

```python
se
```

```python
from scipy.stats import t
p_val=[]
for i in range(13):
    p_temp = 2*(1- t.cdf(abs(mlr.coef_[i]/se[i]), 506-13))
    p_val.append(round(float(p_temp),3))
```
